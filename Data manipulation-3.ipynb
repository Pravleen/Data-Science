#### merging by common key -join and merge operation

```python
d1 = pd.DataFrame({'key': ['K0', 'K8', 'K2', 'K3'],
                     'A': ['A0', 'A1', 'A2', 'A3'],
                     'B': ['B0', 'B1', 'B2', 'B3']})
   
d2 = pd.DataFrame({'key': ['K0', 'K1', 'K2', 'K3'],
                          'C': ['C0', 'C1', 'C2', 'C3'],
                          'D': ['D0', 'D1', 'D2', 'D3']})
```

```python
d1
```

```python
d2
```

```python
merge1=pd.merge(d1,d2) #try to find common column
```

```python
merge1
```

```python
merge1=pd.merge(d1,d2,on="key")
```

```python
merge1
```

```python
pd.merge(d1,d2,on="key",how="inner")
```

```python
pd.merge(d1,d2,on="key",how="outer")
```

```python
pd.merge(d1,d2,on="key",how="left")
```

```python
pd.merge(d1,d2,on="key",how="right")
```

```python
# multikey dataframe
left = pd.DataFrame({'key1': ['K0', 'K0', 'K1', 'K2'],
                     'key2': ['K0', 'K1', 'K0', 'K1'],
                        'A': ['A0', 'A1', 'A2', 'A3'],
                        'B': ['B0', 'B1', 'B2', 'B3']})
    
right = pd.DataFrame({'key1': ['K0', 'K1', 'K1', 'K2'],
                               'key2': ['K0', 'K0', 'K0', 'K0'],
                                  'C': ['C0', 'C1', 'C2', 'C3'],
                                  'D': ['D0', 'D1', 'D2', 'D3']})
```

```python
left
```

```python
right
```

```python
pd.merge(left,right,on=["key1","key2"])
```

```python
#join operators
left = pd.DataFrame({'A': ['A0', 'A1', 'A2'],
                     'B': ['B0', 'B1', 'B2']},
                      index=['K0', 'K1', 'K2']) 

right = pd.DataFrame({'C': ['C0', 'C2', 'C3'],
                    'D': ['D0', 'D2', 'D3']},
                      index=['K0', 'K2', 'K3'])
```

```python
left.join(right)
```

```python
right.join(left)
```

```python
left.join(right,how="outer")
```

```python
left.join(right,how="inner")
```
