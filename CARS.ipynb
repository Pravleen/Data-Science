```python
bins=np.linspace(min(df['mpg']),max(df['mpg']),4)
group_names=["Low","medium","High"]
bins
```

```python
df["mpg-bin"]=pd.cut(df["mpg"],bins,labels=group_names,include_lowest=True)
```
